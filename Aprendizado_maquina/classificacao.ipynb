```python
# !python -m pip install pandas matplotlib seaborn scikit-learn 

import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns  
```

```python
df = pd.read_csv("dados_flu.csv")
df.isnull().sum()
df
```

```python
df_clean = df.drop(["seguro_de_saúde","seguro_de_saúde.1",
                    "indústria_de_emprego", "ocupação",], axis=1)

df_clean = df_clean.dropna()

df_clean.sample(3, random_state=42)
```

```python
corr = df_clean.corr()
fig = plt.figure(figsize=(20,20))
sns.heatmap(corr, cmap="coolwarm", annot=True, center=0)
```

```python
X = df_clean[
    ["opinião_sobre_o_risco_sazonal",
    "preocupação_com_h1n1",
    "opinião_sobre_a_efetividade_da_vacina_sazonal",
    ]]

X
```

```python
y= df_clean["vacina_sazonal"]
y
```

```python
from sklearn.neighbors import KNeighborsClassifier 

predit_vac = KNeighborsClassifier()
predit_vac
```

```python
predit_vac.fit(X,y) # dados tabulares numericos sem nans 
```

```python
#"opinião_sobre_o_risco_sazonal",
#"preocupação_com_h1n1",
#"opinião_sobre_a_efetividade_da_vacina_sazonal",

# lista de pessoas que responderam a essas perguntas  
predit_vac.predict([ # predict recebe a entrada e da a saida  
    [1, 0, 0], # primeira pessoa  
    [3,4,5] # segunda pessoa 
])
```

```python
predit_vac.predict_proba([ # predict proba para saber a probabilidade recebe a entrada e da a saida  
    [1, 0, 0], # primeira pessoa  
    [3,4,5] # segunda pessoa 
])
```
